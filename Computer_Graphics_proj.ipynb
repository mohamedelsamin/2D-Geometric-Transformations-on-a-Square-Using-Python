```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def display_square(points, title="Square"):
    square = np.vstack([points, points[0]])
    plt.figure()
    plt.plot(square[:, 0], square[:, 1], marker='o')
    plt.xlim(-20, 30)
    plt.ylim(-20, 30)
    plt.title(title)
    plt.grid(True)
    plt.gca().set_aspect('equal', adjustable='box')
    plt.show()
```

```python
square_points = np.array([
    [2, 2],
    [6, 2],
    [6, 6],
    [2, 6]
])
display_square(square_points, "Step 1: Original Square")
```

```python
translation_vector = np.array([5, 4])
translated_square = square_points + translation_vector
display_square(translated_square, "Step 2: Translated Square (+5, +4)")
```

```python
theta = np.radians(45)
rotation_matrix = np.array([
    [np.cos(theta), -np.sin(theta)],
    [np.sin(theta),  np.cos(theta)]
])
rotated_square = np.dot(translated_square, rotation_matrix.T)
display_square(rotated_square, "Step 3: Rotated Square (45Â°)")
```

```python
scale_factor = 2
scaled_square = rotated_square * scale_factor
display_square(scaled_square, "Step 4: Scaled Square (x2)")
```
